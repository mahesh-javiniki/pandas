## **11. Customization Options (Basic)**

````python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
````

### üìå 1. What It Does and When to Use

**Customization options** refer to visual modifications you can apply to your time series plots (created using pandas or Matplotlib) to improve readability, highlight patterns, and make your plots aesthetically pleasing and informative.

‚úÖ Use **customization** when:

* You want to tailor your visuals for specific audiences.
* You need to highlight key trends, periods, or comparisons.
* You're preparing plots for reports or dashboards.

## üõ†Ô∏è 2. Syntax and Core Parameters

Most customization is applied using Matplotlib functions that are either directly accessible through `pandas.plot()` or by calling `plt` from `matplotlib.pyplot`.

### General Syntax (for customization)

```python
df.plot(
    title='Title Here',
    xlabel='X Axis Label',
    ylabel='Y Axis Label',
    figsize=(10, 5),
    grid=True,
    color='green',
    linestyle='--',
    marker='o'
)
```

Or using Matplotlib directly:

```python
plt.plot(df.index, df['column'], color='red', linewidth=2, linestyle='-.')
plt.title('Your Title')
plt.xlabel('Date')
plt.ylabel('Values')
plt.grid(True)
plt.xticks(rotation=45)
plt.legend(['Series Name'])
plt.show()
```

## üìä 4. Understanding Each Technique with Examples

````python
# Simulated time series data
date_range = pd.date_range(start='2023-01-01', periods=30)
df = pd.DataFrame({
    'sales': np.random.randint(100, 200, size=30)
}, index=date_range)

df
````

### Example 1: Basic Customized Plot

````python
plt.figure(figsize=(10, 5))
plt.plot(df.index, df['sales'], color='darkblue', linestyle='--', marker='o')
plt.title('Daily Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales Units')
plt.grid(True)
plt.xticks(rotation=45)
plt.legend(['Sales'])
plt.tight_layout()
plt.show()
````

### Example 2: Custom Date Format and Tick Frequency

````python
import matplotlib.dates as mdates

fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(df.index, df['sales'], color='teal', linewidth=2)

# Set major ticks every 5 days and format
ax.xaxis.set_major_locator(mdates.DayLocator(interval=5))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))

plt.title('Sales Trend')
plt.xlabel('Date')
plt.ylabel('Units Sold')
plt.grid(True)
plt.tight_layout()
plt.show()
````

## ‚ö†Ô∏è 5. Common Pitfalls

| Pitfall                        | Why It‚Äôs a Problem                    |
| ------------------------------ | ------------------------------------- |
| ‚ùå Forgetting `tight_layout()`  | Overlaps between labels and plot area |
| ‚ùå Using small figure sizes     | Crowds plot elements                  |
| ‚ùå No label/legend              | Viewers can't interpret the plot      |
| ‚ùå Over-using colors or markers | Makes plot visually noisy             |
| ‚ùå Ignoring date formatting     | Leads to cluttered x-axis             |
