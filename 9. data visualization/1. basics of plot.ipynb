# **9. Data Visualization (with Pandas)**

## **1. Basics of `.plot()` Method**

````python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
````

### 1️⃣ What It Does and When to Use It

The `.plot()` method in pandas allows you to **create basic visualizations directly from Series or DataFrame** objects. It uses **Matplotlib** under the hood and is especially useful for **quick exploratory data analysis (EDA)**.

#### ✅ When to Use:

* During **initial exploration** to understand trends or patterns
* For **quick, inline charts** in notebooks
* When you don't need advanced customizations
* To create **simple visualizations** (line, bar, area, etc.)


### 2️⃣ Syntax and Core Parameters

```python
df.plot(
    kind='line',        # Type of plot: line, bar, barh, hist, box, kde, area, pie, scatter
    x=None,             # Column label for x-axis
    y=None,             # Column label or list of labels for y-axis
    title=None,         # Title of the plot
    figsize=None,       # Tuple (width, height)
    grid=False,         # Show grid lines
    legend=True,        # Show legend
    subplots=False,     # Plot each column in a separate subplot
    colormap=None,      # Color theme for the plots
    sharex=None,        # Share x axis across subplots
    sharey=None,        # Share y axis across subplots
)
```


### 3️⃣ Different Techniques (with Variations)

Pandas’ `.plot()` can be used on:

#### ✅ Series

```python
series.plot()
```

#### ✅ DataFrame (with automatic index for x-axis)

```python
df.plot()                      # Default is line plot
df.plot(kind='bar')           # Bar chart
df.plot(kind='hist')          # Histogram
df.plot(kind='box')           # Box plot
df.plot(kind='area')          # Area chart
df.plot(kind='kde')           # Density plot
```

#### ✅ With Specific X and Y Axes

```python
df.plot(x='col1', y='col2')
```

#### ✅ Multiple columns

```python
df[['col2', 'col3']].plot()
```


### 4️⃣ Understanding Each Technique with Examples (With Data)

````python
data = {
    'Date': pd.date_range(start='2023-01-01', periods=6, freq='ME'),
    'Sales': [200, 220, 250, 300, 280, 330],
    'Profit': [20, 30, 50, 60, 55, 65]
}
df = pd.DataFrame(data)

df
````

#### ✅ 1. Default Line Plot

````python
df.plot(x='Date', y=['Sales', 'Profit'], title='Sales and Profit over time', figsize=(8,4))
````

#### ✅ 2. Column-wise Line Plot (no x-axis specified)

````python
df.set_index('Date').plot(title='Sales and profit trend')
````

#### ✅ 3. Series Plot

````python
df['Sales'].plot(title='Sales over time')
````

````python
df.set_index('Date')['Sales'].plot(title='Sales Over Time')
````

### 5️⃣ Common Pitfalls

| Pitfall                                 | Why it happens                                       | Solution                                  |
| --------------------------------------- | ---------------------------------------------------- | ----------------------------------------- |
| 📉 **Empty Plot**                       | Forgetting `plt.show()` in script-based environments | Add `plt.show()`                          |
| 🔢 **Wrong x-axis**                     | Not setting `x='column'` or indexing properly        | Use `x='Date'` or `.set_index('Date')`    |
| 📛 **Confusing plot types**             | Using `kind='scatter'` without `x` and `y`           | Must specify both `x` and `y` for scatter |
| 🧮 **Large datasets = cluttered plots** | Too many lines or categories                         | Filter data or use subplots               |
| 🎨 **Color mismatch in legends**        | With multiple plots or colormaps                     | Use `colormap='Set1'` or specify manually |


### 6️⃣ Real-World Examples and Use Cases

#### ✅ Business Sales Dashboard (Quick EDA)

````python
# Visualize monthly sales and profit
df.set_index('Date')[['Sales', 'Profit']].plot(title='Monthly Sales vs Profit')
plt.show()
````

#### ✅ Time Series Monitoring

````python
# Assume CPU usage over time
usage = pd.Series(np.random.rand(30)*100, index=pd.date_range('2023-01-01', periods=30))
usage.plot(title='CPU Usage (%) Over Time')
plt.ylabel('Usage (%)')
plt.show()
````

#### ✅ Exploratory Analysis Before Modeling

````python
# Checking correlation visually
df.plot(x='Sales', y='Profit', kind='scatter', title='Sales vs Profit')
plt.show()
````
