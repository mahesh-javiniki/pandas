## üìä 4. **Histogram in pandas**

````python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
````

Histograms are **foundational plots** for understanding **distributions** in your dataset.

### 1Ô∏è‚É£ What It Does and When to Use

A **histogram** is a graphical representation that organizes a group of **continuous numerical data** into **bins (intervals)**. It shows how frequently each range of values occurs.

> In pandas, histograms can be created using either:

* `.hist()` ‚Äî a specialized method just for histograms.
* `.plot(kind='hist')` ‚Äî using the general `.plot()` function.

---

#### ‚úÖ When to Use:

* To analyze **distribution patterns** in numerical data
* To **detect skewness**, **outliers**, and **data spread**
* To **compare the spread** of multiple columns
* To determine **appropriate bin sizes** for modeling or discretization

### 2Ô∏è‚É£ Syntax and Core Parameters

```python
# Using .hist()
df.hist(
    column=None,         # Column(s) to plot
    by=None,             # Group by this column
    bins=10,             # Number of histogram bins
    figsize=None,        # Tuple of figure size
    layout=(r, c),       # Rows and columns of subplots
    sharex=False,        # Share x-axis across subplots
    sharey=False,        # Share y-axis across subplots
    grid=True,           
    color=None,          
)

# Using .plot(kind='hist')
df['col'].plot(
    kind='hist',
    bins=10,
    color='green',
    alpha=0.7,           # Transparency
    edgecolor='black',
    title='Histogram of col'
)
```

### 4Ô∏è‚É£ Understanding Each Technique with Examples (with Data)

````python
np.random.seed(42)
data = {
    'Math': np.random.normal(70, 10, 100),
    'English': np.random.normal(60, 15, 100),
    'Science': np.random.normal(75, 5, 100),
    'Class': np.random.choice(['A', 'B'], size=100)
}
df = pd.DataFrame(data)

df
````

#### üÖ∞Ô∏è A. Simple Histogram using `.hist()`

````python
df['Math'].hist()
````

````python
df['Math'].hist()
plt.title('Marks score histogram')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.show()
````

#### üÖ±Ô∏è B. Specifying Number of Bins

````python
df['Math'].hist(bins=20, color='orange', edgecolor='black')
plt.title('Math scores with 20 bins')
plt.show()
````

#### üÖ≤ C. Multiple Histograms in Subplots

````python
df[['Math', 'English', 'Science']].hist(bins=15, figsize=(10, 5), layout=(1, 3))
plt.title('Subject wise scores distribution')
plt.tight_layout()
plt.show()
````

#### üÖ≥ D. Overlayed Histograms using `.plot(kind='hist')`

````python
df['Math'].plot(kind='hist', bins=15, alpha=0.5, label='Math', legend=True)
df['English'].plot(kind='hist', bins=15, alpha=0.5, label='English', legend=True)

plt.title('Math VS English score distribution')
plt.xlabel('Score')
plt.show()
````

#### üÖ¥ E. Grouped Histograms with `by=`

````python
df.hist(column='Math', by='Class', figsize=(8, 4))
plt.suptitle('Math scores by class')
plt.show()
````

#### üÖµ F. Customized Histogram

````python
df['Science'].plot(kind='hist', bins=12, color='purple', alpha=0.6, edgecolor='black')
plt.title('Science Score Distribution')
plt.grid(True)
plt.show()
````

### 5Ô∏è‚É£ Common Pitfalls

| Pitfall                 | Why it Happens                              | How to Fix                              |
| ----------------------- | ------------------------------------------- | --------------------------------------- |
| ‚ùå **Too few bins**      | May hide important patterns or peaks        | Increase `bins` value                   |
| ‚ùå **Too many bins**     | Leads to noise, sparse bars                 | Reduce `bins` to a meaningful number    |
| ‚ùå **Overlay confusion** | Overlaying multiple histograms can be messy | Use transparency (`alpha`) and legend   |
| ‚ùå **Poor axis labels**  | Plots are unreadable without axis titles    | Always set `xlabel()` and `ylabel()`    |
| ‚ùå **Group not visible** | Using `by=` on numeric column won't help    | Use `by=` on categorical variables only |


### 6Ô∏è‚É£ Real-World Examples and Use Cases

#### üéì Use Case 1: Analyzing Student Performance

````python
df['Math'].hist(bins=15)
plt.title('Distribution of Math Scores')
plt.show()
````

#### üè¢ Use Case 2: Employee Age Distribution in HR Analytics

````python
# Assuming age data
ages = pd.Series(np.random.randint(21, 60, size=100))
ages.plot(kind='hist', bins=10, color='teal', edgecolor='black')
plt.title('Employee Age Distribution')
plt.xlabel('Age')
plt.show()
````

#### üíµ Use Case 3: Income Distribution in Finance

````python
# Simulate income
incomes = pd.Series(np.random.lognormal(mean=10, sigma=0.5, size=5000))
incomes.plot(kind='hist', bins=50, log=True, color='coral', title='Income Distribution (log scale)')
plt.xlabel('Income')
plt.show()
````
