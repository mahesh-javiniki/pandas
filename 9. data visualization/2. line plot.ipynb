# **9. Data Visualization (with Pandas)**

## üìà 2. **Line Plot in pandas**

````python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
````

### 1Ô∏è‚É£ What It Does and When to Use

A **line plot** is a graph that displays information as a series of **data points connected by straight line segments**. In pandas, this is the **default plot type** when calling `.plot()` on a DataFrame or Series.

#### ‚úÖ When to Use:

* When visualizing **time series data**
* To analyze **trends over time** (e.g., monthly sales, daily temperatures)
* When comparing **multiple metrics across the same index**
* For **continuous data** where relationships between points matter


### 2Ô∏è‚É£ Syntax and Core Parameters

```python
df.plot(
    kind='line',       # (Optional) Line plot is the default
    x=None,            # Column for x-axis
    y=None,            # Column(s) for y-axis
    title=None,        # Title of the plot
    figsize=None,      # Figure size in inches (width, height)
    linestyle='-',     # Line style: '-', '--', '-.', ':'
    marker=None,       # Markers: 'o', 's', '^', etc.
    color=None,        # Line color
    legend=True,       # Show legend
    grid=False,        # Show grid
)
```


### 3Ô∏è‚É£ Different Techniques (with Variations)

We‚Äôll explore the following variations of line plots:

| Technique                         | Description                                |
| --------------------------------- | ------------------------------------------ |
| A. Default line plot              | Basic plot using DataFrame index           |
| B. Plot with explicit `x` and `y` | Specify x-axis and y-axis columns          |
| C. Multiple lines                 | Compare multiple numeric columns           |
| D. Custom styles                  | Change line style, marker, color           |
| E. Subplots                       | Create separate line plots for each column |
| F. Plot Series directly           | Line plot from a single column             |


### 4Ô∏è‚É£ Understanding Each Technique with Examples

````python
data = {
    'Date': pd.date_range(start='2024-01-01', periods=6, freq='ME'),
    'Sales': [200, 220, 250, 300, 280, 330],
    'Profit': [20, 30, 50, 60, 55, 65]
}
df = pd.DataFrame(data)

df
````

#### üÖ∞Ô∏è A. Default Line Plot (index used as x-axis)

````python
df.set_index('Date')[['Sales', 'Profit']].plot(title='Default Line Plot')
````

#### üÖ±Ô∏è B. Line Plot with `x` and `y`

````python
df.plot(x='Date', y='Sales', title='Sales over time')
plt.ylabel('Sales ($)')
````

#### üÖ≤ C. Multiple Lines

````python
df.plot(x='Date', y=['Sales', 'Profit'], title='Sales vs Profit over time')
````

#### üÖ≥ D. Custom Styles

````python
df.plot(
    x='Date', y='Sales', 
    linestyle='--', marker='o', color='green', 
    title='Custom Styled Sales'
)
````

#### üÖ¥ E. Subplots for Each Column

````python
df.set_index('Date').plot(subplots=True, layout=(2, 1),
                          figsize=(8, 6), title='Subplots of Sales and Profit')
plt.tight_layout()
````

#### üÖµ F. Plotting Series (single column)

````python
df.set_index('Date')['Sales'].plot(title='Sales line plot')
plt.ylabel('Sales')
plt.show()
````

### 5Ô∏è‚É£ Common Pitfalls

| Pitfall                                  | Explanation                                | Fix                                    |
| ---------------------------------------- | ------------------------------------------ | -------------------------------------- |
| ‚ùå **Using string dates without parsing** | Plotting string dates will not work well   | Use `pd.to_datetime()`                 |
| ‚ùå **Missing `plt.show()` in scripts**    | No plot will be displayed                  | Always call `plt.show()` in scripts    |
| ‚ùå **Too many lines clutter the plot**    | Hard to read when many columns are plotted | Limit to 2‚Äì3 lines or use subplots     |
| ‚ùå **Index not set as date**              | x-axis looks like row numbers              | Use `.set_index('Date')` or `x='Date'` |
| ‚ùå **Inconsistent time intervals**        | Makes line chart misleading                | Resample or interpolate missing dates  |


### 6Ô∏è‚É£ Real-World Examples and Use Cases

#### üìä Use Case 1: Visualizing Monthly Revenue Growth

````python
df.set_index('Date')['Sales'].plot(title='Monthly Revenue')
plt.ylabel('Revenue ($)')
plt.grid(True)
plt.show()
````

#### üí∞ Use Case 2: Comparing Sales and Profit Trends

````python
df.set_index('Date')[['Sales', 'Profit']].plot(title='Sales vs Profit Trend')
plt.ylabel('Value ($)')
plt.show()
````

#### üìà Use Case 3: Monitoring Metrics Over Time

````python
cpu_usage = pd.Series(np.random.rand(30)*100, index=pd.date_range('2024-01-01', periods=30))
cpu_usage.plot(title='CPU Usage Over 30 Days')
plt.ylabel('CPU %')
plt.grid(axis='both')
plt.show()
````
