# **9. Data Visualization (with Pandas)**

## **8. Scatter Plot**

````python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
````

## 1. ‚úÖ What It Does and When to Use

A **scatter plot** visualizes the **relationship between two numerical variables**. In the context of time series or datetime-based data, scatter plots can show:

* How one variable changes over time (e.g., temperature vs time).
* Correlation between two variables (e.g., sales vs profit).
* Irregularly spaced data over time.

> üîé **Use Case:** When you want to detect trends, correlations, clusters, or outliers between variables ‚Äî especially in time-stamped datasets.

## 2. üß© Syntax and Core Parameters

```python
df.plot.scatter(x='column_x', y='column_y', s=marker_size, c=color, colormap='viridis', alpha=0.8)
```

### Key Parameters:

| Parameter                   | Description                                      |
| --------------------------- | ------------------------------------------------ |
| `x`                         | Column for x-axis values (e.g., datetime)        |
| `y`                         | Column for y-axis values                         |
| `s`                         | Marker size (int or array)                       |
| `c`                         | Marker color (can be array or color name)        |
| `colormap`                  | Apply color gradients (useful when `c` is array) |
| `alpha`                     | Transparency (0 to 1)                            |
| `title`, `xlabel`, `ylabel` | Plot title and axis labels                       |


## 3. üîÄ Techniques with Variations

| Technique                 | Description                                            |
| ------------------------- | ------------------------------------------------------ |
| Basic scatter plot        | x vs y                                                 |
| Color by category         | Use color array for coloring points                    |
| Size by value             | Use array for size (`s`)                               |
| Time on x-axis            | Use datetime column on x-axis                          |
| Colored by third variable | Add more insights by coloring points by another metric |


## 4. üß™ Understanding with Examples

````python
# Create sample datetime and metrics
np.random.seed(0)
dates = pd.date_range(start="2023-01-01", periods=100, freq='D')
df = pd.DataFrame({
    'date': dates,
    'sales': np.random.normal(200, 25, size=100),
    'profit': np.random.normal(50, 10, size=100),
    'units_sold': np.random.randint(10, 100, size=100)
})

df
````

### 1Ô∏è‚É£ Basic Scatter Plot (Time vs Sales)

````python
df.plot(kind='scatter', x='date', y='sales', title="Sales over time", alpha=0.7)
plt.show()
````

````python
df.plot.scatter(x='date', y='sales', title='Sales Over Time', alpha=0.7)
plt.show()
````

> Useful for visualizing fluctuations and gaps in sales over time.

### 2Ô∏è‚É£ Scatter Plot with Color (Color by Profit)

````python
df.plot.scatter(x='date', y='sales', c='profit', colormap='coolwarm', title='Sales colored by Profit')
plt.show()
````

> Adds third dimension (profit) using color gradients.

### 3Ô∏è‚É£ Scatter with Size Encoding (Units Sold as Size)

````python
df.plot.scatter(x='date', y='sales', s=df['units_sold'], title='Sales with Size = Units Sold', alpha=0.6)
plt.show()
````

> Gives an idea of volume sold on each day.

### 4Ô∏è‚É£ Correlation Between Two Variables (Sales vs Profit)

````python
df.plot.scatter(x='sales', y='profit', title='Sales vs Profit', alpha=0.7)
plt.show()
````

> Quickly helps spot correlations or anomalies between two metrics.

## 5. ‚ö†Ô∏è Common Pitfalls

| Pitfall                               | Explanation                                        |
| ------------------------------------- | -------------------------------------------------- |
| ‚ùå Plotting too many points            | Causes clutter ‚Äî consider transparency or sampling |
| ‚ùå Using too large markers             | Overlaps may mislead interpretation                |
| ‚ùå Misaligned datetime on x-axis       | Always check datetime column is in datetime format |
| ‚ùå Ignoring outliers                   | Can distort scale or patterns                      |
| ‚ùå Mixing scales without normalization | Make sure units are consistent or scaled           |
