# **9. Data Visualization (with Pandas)**

## **6. Area Plot**

````python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
````

## ✅ 1. What It Does and When to Use

### 🔹 What:

An **area plot** is a variation of a line plot where the area between the axis and the line is filled with color. It’s mainly used to show **quantitative data over time**, emphasizing the **magnitude of change**.

By default, pandas' `DataFrame.plot(kind='area')` generates stacked area plots, which are helpful for visualizing **part-to-whole relationships** over time.

### 🔹 When to Use:

* To show **trend and volume** over time
* To compare **contributions of multiple variables** to a whole
* To visualize **cumulative data**
* When you want to emphasize the **size** of quantities, not just direction

## ✅ 2. Syntax and Core Parameters

```python
df.plot(kind='area', stacked=True, alpha=0.5, figsize=(10, 5))
```

### 📌 Core Parameters:

| Parameter     | Description                                         |
| ------------- | --------------------------------------------------- |
| `kind='area'` | Specifies that you want an area plot                |
| `stacked`     | Whether to stack the area plots (`True` by default) |
| `alpha`       | Opacity of the fill (0 transparent → 1 opaque)      |
| `x`           | Label or position of the x-axis column              |
| `y`           | List of columns to plot on the y-axis               |
| `figsize`     | Tuple specifying figure size                        |
| `title`       | Title of the plot                                   |
| `grid`        | Add a grid (default: `False`)                       |
| `color`       | Custom color for each series                        |


## ✅ 3. Techniques and Variations

### 🔸 Basic Area Plot

Plotting a single column or multiple columns with default stacking.

### 🔸 Unstacked Area Plot

Set `stacked=False` to avoid overlapping.

### 🔸 Transparency Control

Use `alpha` to enhance readability when multiple areas are plotted.

### 🔸 Area Plot for Time Series

Works well with datetime index for visualizing trends over time.


## ✅ 4. Examples with Data

````python
# Create dummy sales data
data = {
    'Month': pd.date_range('2024-01-01', periods=6, freq='ME'),
    'Product_A': [120, 150, 170, 200, 240, 300],
    'Product_B': [100, 130, 160, 180, 200, 220],
    'Product_C': [80, 100, 120, 140, 160, 180]
}

df = pd.DataFrame(data).set_index('Month')

df
````

### 🔹 1. Basic Area Plot (Stacked - Default)

````python
df.plot(kind='area', title='Monthly Sales', figsize=(10, 5))
plt.show()
````

### 🔹 2. Unstacked Area Plot

````python
df.plot(kind='area', stacked=False, title="Unstacked Area Plot", figsize=(10,5))
plt.show()
````

### 🔹 3. Area Plot with Transparency

````python
df.plot(kind='area', alpha=0.4, title="Area Plot with Transparency", figsize=(10,5))
plt.show()
````

### 🔹 4. Area Plot for Specific Columns

````python
df[['Product_A', 'Product_B']].plot(kind='area', stacked=True, title="Selected Products", figsize=(10,5))
plt.show()
````

## ✅ 5. Common Pitfalls

| Pitfall                     | Explanation                                                                                                                                                      |
| --------------------------- | ---------------------------------------------------------------------------------------------------------------------------------------------------------------- |
| **Too many categories**     | Makes the area chart cluttered and hard to interpret. Stick to 3–5 layers ideally.                                                                               |
| **Stacking confusion**      | When `stacked=True`, the cumulative nature may mislead viewers into assuming actual values are increasing. Use `stacked=False` for clarity in individual trends. |
| **Opacity issues**          | Without adjusting `alpha`, overlapping layers may become unreadable.                                                                                             |
| **Improper x-axis**         | Ensure datetime or ordered categories for meaningful trend visualization.                                                                                        |
| **Plotting on wrong index** | Always set the datetime column as index for time series visualization.                                                                                           |
