## **üìä 3. Bar Plot in pandas**

````python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
````

> Bar plots are another essential tool in EDA (Exploratory Data Analysis), especially when you're working with categorical or comparative data.

### 1Ô∏è‚É£ What It Does and When to Use

A **bar plot** (also known as a bar chart) represents data with **rectangular bars** with lengths proportional to the values they represent. In pandas, this is specified using `kind='bar'` or `kind='barh'` for horizontal bar plots.

#### ‚úÖ When to Use:

* To compare **quantitative values** across **categories**
* When dealing with **aggregated statistics** like sum, mean, or count
* To visualize **discrete variables** (e.g., sales by region, counts by category)
* For **grouped** or **stacked** visual comparison of multiple variables


### 2Ô∏è‚É£ Syntax and Core Parameters

```python
df.plot(
    kind='bar',        # or 'barh' for horizontal bars
    x=None,            # Column name for x-axis (optional if index is meaningful)
    y=None,            # Column(s) for y-axis
    stacked=False,     # Stacked bar chart or grouped
    color=None,        # Color of bars
    title=None,        # Title of plot
    figsize=None,      # Size in inches (width, height)
    rot=0,             # Rotation of x-axis labels
    width=0.8,         # Width of each bar
    legend=True,       # Display legend
)
```

### 3Ô∏è‚É£ Different Techniques (with Variations)

| Technique                        | Description                            |
| -------------------------------- | -------------------------------------- |
| A. Basic vertical bar plot       | Standard `kind='bar'`                  |
| B. Horizontal bar plot           | `kind='barh'`                          |
| C. Grouped bar plot              | Multiple values per category           |
| D. Stacked bar plot              | `stacked=True`                         |
| E. Bar plot from aggregated data | With `.groupby()` or `.value_counts()` |
| F. Customized plot               | Colors, rotation, width, etc.          |


### 4Ô∏è‚É£ Understanding Each Technique with Examples

````python
data = {
    'Product': ['A', 'B', 'C', 'D'],
    'Q1_Sales': [100, 150, 80, 120],
    'Q2_Sales': [90, 160, 70, 130]
}
df = pd.DataFrame(data)

df
````

#### üÖ∞Ô∏è A. Basic Vertical Bar Plot

````python
df.plot(kind='bar', x='Product', y='Q1_Sales', title='Q1 Sales by Product')
plt.ylabel('Sales')
plt.show()
````

#### üÖ±Ô∏è B. Horizontal Bar Plot

````python
df.plot(kind='barh', x='Product', y='Q1_Sales', title='Q1 Sales by Product')
plt.ylabel('Sales')
plt.show()
````

#### üÖ≤ C. Grouped Bar Plot (Compare Two Quarters)

````python
df.plot(kind='bar', x='Product', y=['Q1_Sales', 'Q2_Sales'], title='Sales by product')
plt.ylabel('Sales')
plt.show()
````

#### üÖ≥ D. Stacked Bar Plot

````python
df.plot(kind='bar', x='Product', y=['Q1_Sales', 'Q2_Sales'], stacked=True, title='Stacked Quarterly Sales')
plt.ylabel('Sales')
plt.show()
````

#### üÖ¥ E. Bar Plot from Aggregated Data

````python
sales_data = pd.DataFrame({
    'Category': ['Electronics', 'Furniture', 'Clothing', 'Clothing', 'Furniture', 'Electronics'],
    'Sales': [100, 200, 150, 130, 220, 90]
})

sales_data
````

````python
category_sales = sales_data.groupby('Category')['Sales'].sum()

display(category_sales)
category_sales.plot(kind='bar', title='Total Sales by Category')
plt.ylabel('Sales')
plt.show()
````

#### üÖµ F. Customized Bar Plot

````python
df.plot(
    kind='bar', 
    x='Product', y='Q1_Sales', 
    color='green', width=0.5,
    title='Customized Q1 Sales',
    rot=45,
    figsize=(6, 4)
)
plt.ylabel('Sales')
plt.grid(True)
plt.show()
````

### 5Ô∏è‚É£ Common Pitfalls

| Pitfall                                  | Explanation                                    | Fix                                       |
| ---------------------------------------- | ---------------------------------------------- | ----------------------------------------- |
| ‚ùå **Bar order confusion**                | Index order may not reflect natural sorting    | Use `.sort_values()` if needed            |
| ‚ùå **Overlapping labels**                 | Long category names may clash                  | Use `rot=45` or `figsize=(width, height)` |
| ‚ùå **Missing legend in grouped bar**      | Legends may be unclear when multiple y columns | Ensure `legend=True`                      |
| ‚ùå **Using bar plot for continuous data** | Bars are for categories, not continuous time   | Use **line plots** for trends             |
| ‚ùå **Cluttered stacked bars**             | Too many components make it unreadable         | Limit to 2‚Äì3 components per stack         |


### 6Ô∏è‚É£ Real-World Examples and Use Cases

#### üì¶ Use Case 1: Product Category Sales

````python
sales_data
````

````python
df.plot(kind='bar', x='Product', y=['Q1_Sales', 'Q2_Sales'], title='Quarterly Comparison')
plt.ylabel('Sales')
plt.show()
````

#### üèÜ Use Case 3: Top Performers in a Team

````python
team_data = pd.Series([80, 95, 70, 85], index=['Alice', 'Bob', 'Carol', 'David'])
team_data.plot(kind='bar', title='Employee Performance Score')
plt.ylabel('Score')
plt.show()
````

| **Visual Variants** | Grouped, stacked, colored, horizontal                              |
