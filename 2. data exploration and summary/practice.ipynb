## 4. Inference & Conversion Utilities

```python
df = pd.DataFrame({
    'bool_column': [True, False, True, True, False]
})

# Convert using pd.to_numeric
df['int_column'] = pd.to_numeric(df['bool_column'])

print(df)
```

```python
df = pd.DataFrame({
    'bool_column': [True, False, True, True, False]
})

df['int_column'] = df['bool_column'].astype(int)

print(df)
print(df.dtypes)
```

```python
titanic.alone = titanic.alone.astype('int')
```

```python
age_median = titanic.age.median(skipna=True)
print(f"Age median: {age_median}")


updated_age_col = titanic.age.fillna(age_median)
updated_age_col
```

## Group Based Summary

```python
titanic.head()
```

```python
titanic.groupby('survived')['age'].median()
```

```python
titanic.groupby('survived')['age'].agg(['min', 'max', 'mean'])
```

```python
titanic['average_age'] = titanic.groupby('survived')['age'].transform('mean')
```

```python
titanic.head()
```

```python
titanic.pivot_table(
    index='survived',
    values=['age', 'alone'],
    aggfunc={'age': 'mean', 'alone': 'max'}
)
```
