## Practice with Titanic dataset

```python
import pandas as pd
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

## **1. Basic Selection Techniques**

```python
titanic['survived'].head()
```

```python
titanic[['survived', 'sex', 'age']].head()
```

```python
titanic.loc[1]
```

```python
titanic.loc[[1, 3]]
```

```python
titanic.iloc[1]
```

```python
titanic.iloc[[1, 3]]
```

```python
# slicing
titanic[1: 5]
```

```python
titanic[1: 5: 2]
```

```python
titanic.loc[1: 5, 'survived']
```

```python
titanic.loc[1: 5, 'survived': 'fare']
```

```python
titanic.loc[1: 5, ['survived', 'fare']]
```

## 2. **Indexing and Reindexing**

```python
titanic.head()
```

```python
titanic1 = titanic.set_index('survived')
titanic1.head()
```

```python
titanic1.loc[1]
```

```python
titanic1.reset_index()
```

```python
titanic.head()
```

```python
titanic.index = titanic['survived']
```

```python
titanic.head()
```

```python
titanic.reset_index(drop=True)
```

```python
# Hierarchical indexing

titanic.set_index(['sex', 'survived'])
```
